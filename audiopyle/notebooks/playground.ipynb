```python
from audiopyle.audio import Audio
import librosa
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import Normalize
```

```python
# Compute Mel Frequence Cepstral Coefficients (MFCCs) from an audio file
filepath = "/workspaces/audiopyle/audiopyle/audiopyle/data/Lars Huismann - Sounds From The Past III - 01 Sounds From The Past III.mp3"
audio = Audio._from_filepath(filepath)

# sr=None implies the native sampling rate, which is twice the reccomended
y, sample_rate = librosa.load(audio._filepath, sr=None)
```

```python
n_mfcc = 20 # number of MFCCs to compute (Usually 12-13)
```

```python
# Setup Plot
fig, ax = plt.subplots(nrows=2, figsize=(10, 8), sharex=True)

# Generate Mel-Frequence Cepstral Coefficients (MFCCs)
mfcc_groove = np.array(librosa.feature.mfcc(y=y, sr=sample_rate, n_mfcc=n_mfcc))

# Generate Mel Spectogram
S = librosa.feature.melspectrogram(y=y, sr=sample_rate, n_mels=128)


# Visualize mel spectrogram
# Darker - Quiter
# Lighter - Louder
img = librosa.display.specshow(librosa.power_to_db(S, ref=np.max),
                               x_axis='time', y_axis='mel', ax=ax[0], sr=sample_rate)

fig.colorbar(img, ax=[ax[0]], format='%+2.0f dB')
ax[0].set(title='Mel Spectrogram - Sounds From The Past III')
ax[0].label_outer()

# Visualize MFCCs
# Redder - More prominent
# Bluer - Less Prominent
img = librosa.display.specshow(mfcc_groove, x_axis='time', ax=ax[1], sr=sample_rate)
fig.colorbar(img, ax=[ax[1]])
ax[1].set(title="Mel-Frequency Cepstral Coefficients - Sounds From The Past III")
ax[1].set_ylabel("Mel-Frequency Cepstral Coefficient")

for mel in range(n_mfcc):
    ax[1].axhline(mel+.5, color='black', linewidth=1)
```
